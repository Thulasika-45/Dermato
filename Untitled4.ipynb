```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import numpy as np
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt

# Load the trained model
model = load_model(r"C:\Users\thulasika\OneDrive\Music\dermato dataset\Skin_Disease_Classifier.h5")
print("Model Loaded Successfully")

# Ensure model is compiled
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Define the class labels with descriptions and recommended diagnosis
class_labels = {
    0: ("Acne", "A common skin condition causing pimples, blackheads, and cysts.", 
        "Recommended Diagnosis: Can be treated with topical treatments, oral medications, and proper skincare."),
    1: ("Actinic Cheilitis", "A precancerous condition affecting the lips due to sun damage.", 
        "Recommended Diagnosis: Early treatment includes sun protection and topical medications."),
    2: ("Actinic Keratosis", "A rough, scaly patch on the skin caused by sun exposure.", 
        "Recommended Diagnosis: May require cryotherapy, topical creams, or laser therapy."),
    3: ("Atopic Dermatitis", "A chronic skin condition causing dry, itchy, and inflamed skin.", 
        "Recommended Diagnosis: Managed with moisturizers, antihistamines, and steroid creams."),
    4: ("Drug Eruptions", "Skin reactions caused by medications.", 
        "Recommended Diagnosis: Treatment involves discontinuing the drug and using anti-inflammatory medications."),
    5: ("Eczema", "A condition causing dry, itchy, and inflamed skin patches.", 
        "Recommended Diagnosis: Managed with moisturizers, steroids, and lifestyle changes."),
    6: ("Exanthems", "Widespread rashes caused by viral or bacterial infections.", 
        "Recommended Diagnosis: Usually self-limiting but may require symptomatic treatment."),
    7: ("Herpes", "A viral infection causing painful blisters.", 
        "Recommended Diagnosis: Treated with antiviral medications."),
    8: ("Melanoma", "A serious type of skin cancer originating in pigment cells.", 
        "Recommended Diagnosis: Early detection is critical; may require surgery, chemotherapy, or immunotherapy."),
    9: ("Pemphigus", "A rare autoimmune disorder causing blistering skin lesions.", 
        "Recommended Diagnosis: Managed with immunosuppressive drugs."),
    10: ("Psoriasis", "An autoimmune disease causing red, scaly skin patches.", 
        "Recommended Diagnosis: Managed with topical treatments, biologics, and light therapy."),
    11: ("Rosacea", "A chronic skin condition causing facial redness and swelling.", 
        "Recommended Diagnosis: Managed with topical and oral medications."),
    12: ("Sun-Damaged Skin", "Skin changes due to prolonged sun exposure.", 
        "Recommended Diagnosis: Preventive measures include sunscreen and antioxidant skincare."),
    13: ("Tinea (Ringworm) / Candidiasis", "Fungal infections causing itchy, ring-shaped rashes.", 
        "Recommended Diagnosis: Treated with antifungal creams or oral antifungal medications."),
    14: ("Vitiligo", "A condition causing loss of skin pigment in patches.", 
        "Recommended Diagnosis: Managed with phototherapy and topical treatments."),
    15: ("Warts", "Small, rough growths caused by HPV.", 
        "Recommended Diagnosis: Treated with cryotherapy, salicylic acid, or laser treatment."),
}

# Function to preprocess a single image
def load_and_preprocess_image(img_path, target_size=(224, 224)):
    img = image.load_img(img_path, target_size=target_size)  # Load image
    img_array = image.img_to_array(img)  # Convert to array
    img_array = np.expand_dims(img_array, axis=0)  # Expand dimensions for model
    img_array /= 255.0  # Normalize the image
    
    # Display the preprocessed image
    plt.figure()
    plt.imshow(img_array[0])  # Display the normalized image
    plt.axis("off")
    plt.title("Preprocessed Image")
    plt.show()
    
    return img_array

# Specify the path of a single test image
test_image_path = r"C:\Users\thulasika\OneDrive\Music\dermato dataset\classified dataset (train)\vitiligo\vitiligo-7.jpg"

# Preprocess the image
img_array = load_and_preprocess_image(test_image_path)

# Make prediction
predictions = model.predict(img_array)
predicted_class_index = np.argmax(predictions)  # Get the class index with the highest probability
predicted_label, description, diagnosis = class_labels.get(predicted_class_index, 
    ("Unknown Disease", "No description available.", "Recommended Diagnosis: No diagnosis available."))



# Print prediction results with separate lines
print(f"Predicted Disease: {predicted_label}\n")
print(f"Description: {description}\n")
print(f"{diagnosis}\n")
```
